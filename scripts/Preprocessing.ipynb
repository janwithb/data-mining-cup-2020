```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.neural_network import MLPRegressor
```

```python
infos = pd.read_csv('../data/infos.csv', sep='|', 
                    converters = {'itemID': str})
items = pd.read_csv('../data/items.csv', sep='|', 
                    converters = {'itemID': str, 'brand': str, 'manufacturer': str, 
                                  'category1': str, 'category2': str, 'category3': str})
orders = pd.read_csv('../data/orders.csv', sep='|', 
                     converters = {'time': np.datetime64, 'transactID': str, 'itemID': str})
```

```python
orders['totalSalesPrice'] = orders['order']*orders['salesPrice']
```

```python
orders_per_day = orders.copy()
```

```python
orders_per_day['time'] = orders_per_day['time'].dt.date
```

```python
orders_per_day.drop(['salesPrice', 'transactID'], axis=1, inplace=True)
```

```python
orders_per_day = orders_per_day.groupby(['time', 'itemID']).sum().reset_index()
```

```python
orders_per_day['salesPrice'] = orders_per_day['totalSalesPrice'] / orders_per_day['order']
```

```python
# Feature Engineering Promotion Feature
from tqdm.notebook import tqdm
for item in tqdm(orders_per_day['itemID'].unique()): 
    df = orders_per_day.loc[orders_per_day['itemID']==item, ['time', 'order']]
    mean = df['order'].sum()/(179)
    std = df['order'].std()
    orders_per_day.loc[(orders_per_day['itemID']==item), 'promotion'] =  [int(x > (mean+0.2*std)) for x in df['order']]
```

```python
df_list = []
item_ids = orders['itemID'].unique()
for date in np.arange(np.datetime64('2018-01-01'), np.datetime64('2018-06-30')):
    df_ = pd.DataFrame({'itemID': item_ids})
    df_['time'] = date
    df_list.append(df_)
```

```python
daily_orders = pd.concat(df_list)
daily_orders['time'] = daily_orders['time'].dt.date
```

```python
data = pd.merge(left=daily_orders, right=orders_per_day, how='left', left_on=['itemID', 'time'], right_on=['itemID', 'time'])
```

```python
data['day'] = (data['time'].astype(np.datetime64) - np.datetime64('2018-01-01')).dt.days
```

```python
data['weekDay'] = data['time'].astype(np.datetime64).dt.weekday
```

```python
data = pd.merge(left=data, right=items, how='left', left_on=['itemID'], right_on=['itemID'])
```

```python
data[data['itemID']=='1']
```

```python
from tqdm.notebook import tqdm
data['order'].fillna(value = 0, inplace=True)
data['totalSalesPrice'].fillna(value = 0, inplace=True)
data['promotion'].fillna(value = 0, inplace=True)
# Dauert 2 Stunden... 
# for item in tqdm(data['itemID'].unique()): 
#     data.loc[(data['itemID']==item), 'salesPrice'] = data.loc[(data['itemID']==item), 'salesPrice'].fillna(method='ffill').fillna(method='bfill')
data['salesPrice'].fillna(value = -1, inplace = True)
```

## Feature Engineering

```python
data = pd.merge(left=data, right=data.groupby(['category2', 'day']).sum()['promotion'].reset_index(), how='left', left_on=['category2', 'day'], right_on=['category2', 'day'], suffixes=('', 'Cat2'))
data = pd.merge(left=data, right=data.groupby(['manufacturer', 'day']).sum()['promotion'].reset_index(), how='left', left_on=['manufacturer', 'day'], right_on=['manufacturer', 'day'], suffixes=('', 'Manu')) 
data = pd.merge(left=data, right=data.groupby(['brand', 'day']).sum()['promotion'].reset_index(), how='left', left_on=['brand', 'day'], right_on=['brand', 'day'], suffixes=('', 'Brand')) 
```

```python
data = data.rename(columns={'time': 'timestamp', 'day': 'time'})
data
```

## Modify prediction set

```python
item_dfs = []
for item_id in tqdm(infos['itemID'].unique()):
    promotions = infos[infos['itemID'] == item_id]['promotion']
    if promotions.isnull().values[0] == True:
        promotions = []
    else:
        promotions = promotions.values[0].split(',')
    day_range = pd.date_range('2018-06-30', periods=14, freq='D')
    item_df = pd.DataFrame({'time': day_range})
    item_df['weekDay'] = item_df['time'].dt.weekday
    item_df.loc[item_df['time'].isin(promotions),'promotion'] = 1
    item_df.loc[~item_df['time'].isin(promotions),'promotion'] = 0
    item_df['promotion'] = item_df['promotion'].astype(int)
    item_df['time'] = np.arange(160, 174)
    item_df['itemID'] = item_id
    item_dfs.append(item_df) 
    
```

```python
data_pred = pd.concat(item_dfs)
```

```python
data_pred = pd.merge(left=data_pred, right=items, how='left', left_on=['itemID'], right_on=['itemID'])
```

```python
data_pred = pd.merge(left=data_pred, right=data_pred.groupby(['category2', 'time']).sum()['promotion'].reset_index(), how='left', left_on=['category2', 'time'], right_on=['category2', 'time'], suffixes=('', 'Cat2'))
data_pred = pd.merge(left=data_pred, right=data_pred.groupby(['manufacturer', 'time']).sum()['promotion'].reset_index(), how='left', left_on=['manufacturer', 'time'], right_on=['manufacturer', 'time'], suffixes=('', 'Manu')) 
data_pred = pd.merge(left=data_pred, right=data_pred.groupby(['brand', 'time']).sum()['promotion'].reset_index(), how='left', left_on=['brand', 'time'], right_on=['brand', 'time'], suffixes=('', 'Brand')) 
```

# Prediction

```python
def get_prediction(model, X_pred):
    prediction = model.predict(X_pred)
    prediction = [int(y) for y in prediction]
    prediction = max(0, int(np.array(prediction).sum()))
    return prediction
```

```python
def train_model(X_train, y_train):
    nn = MLPRegressor(hidden_layer_sizes=(512, 256, 256), max_iter=1000)
    model = nn.fit(X_train, y_train)
    return model
```

```python
# main loop

print('---- Start Model Prediction ----')
result_df = pd.DataFrame(columns=['itemID', 'demandPrediction'])
for item_id in tqdm(data['itemID'].unique()[:100]):
    
    # selecting item 
    item = data[data['itemID']==item_id].reset_index(drop=True)
    item_pred = data_pred[data_pred['itemID']==item_id].reset_index(drop=True)
    
    # create training dataset
#     X_train = item[['time', 'promotion', 'weekDay', 'promotionCat2', 'promotionManu', 'promotionBrand']]
    X_train = item[['time', 'promotion', 'weekDay']]
    y_train = item['order']
        
    print('Training model for item: ', item_id)
    # train model with all data
    model = train_model(X_train, y_train)
    
    # create test dataset
#     X_pred = item_pred[['time', 'promotion', 'weekDay', 'promotionCat2', 'promotionManu', 'promotionBrand']]
    X_pred = item_pred[['time', 'promotion', 'weekDay']]

    
    print('Predicting orders for item: ', item_id)
    # get prediction
    prediction = get_prediction(model, X_pred)
    
    # append prediction to result dataframe
    result_df = result_df.append({'itemID': item_id, 'demandPrediction': prediction}, ignore_index=True)
    
    print('\n')

print('---- End of Model Prediction ----')
```

```python
# convert result dataframe to output format
result_df.set_index('itemID', inplace=True)
index = pd.RangeIndex(1, 10464)
result_df = result_df.reindex(index).fillna(0)
result_df = result_df.reset_index()
result_df = result_df.rename(columns={'index': 'itemID'})
```

## Evaluation

```python
def scoring_function(pred, true, priceData, verbose=0): 
    priceData = priceData
    score = 0
    for itemID in true['itemID'].unique(): 
        s = 0
        actual = true[true['itemID']==itemID]['demandPrediction'].iloc[0]
        predicted = pred[pred['itemID']==itemID]['demandPrediction'].iloc[0]
        price = priceData[priceData['itemID']==itemID]['simulationPrice'].iloc[0]
        if actual >= predicted: 
            s += price * predicted
        else: 
            s += price * actual
            s -= 0.6 * price * (predicted - actual)
        if verbose > 0: 
            print('Score for item with ID {}:\t {:.2f} (max: {})'.format(itemID, s, price*actual))
        score += s
    return score
```

```python
# main loop

print('---- Start Model Prediction ----')
result_df = pd.DataFrame(columns=['itemID', 'demandPrediction'])
real_df = pd.DataFrame(columns=['itemID', 'demandPrediction'])
for item_id in tqdm(data['itemID'].unique()[:500]):
    
    # selecting item 
    item = data[data['itemID']==item_id].reset_index(drop=True)
    item_pred = data_pred[data_pred['itemID']==item_id].reset_index(drop=True)
    
    # create training dataset
    X_train = item[['time', 'promotion', 'weekDay']].iloc[:-14]
    y_train = item['order'].iloc[:-14]
        
    print('Training model for item: ', item_id)
    # train model with all data
    model = train_model(X_train, y_train)
    
    # create test dataset
#     X_pred = item_pred[['time', 'promotion', 'weekDay', 'promotionCat2', 'promotionManu', 'promotionBrand']]
    X_pred = item[['time', 'promotion', 'weekDay']].iloc[-14:]
    y_pred = item['order'].iloc[-14:].sum()
    
    print('Predicting orders for item: ', item_id)
    # get prediction
    prediction = get_prediction(model, X_pred)
    
    # append prediction to result dataframe
    result_df = result_df.append({'itemID': item_id, 'demandPrediction': prediction}, ignore_index=True)
    real_df = real_df.append({'itemID': item_id, 'demandPrediction': y_pred}, ignore_index=True)
    print('\n')
        
print('---- End of Model Prediction ----')
scoring_function(result_df, real_df, infos, verbose=1)
```

```python
def calc_scoring(pred, true, priceData, verbose=0): 
    priceData = priceData
    score = 0
    score_list = []
    for itemID in true['itemID'].unique(): 
        s = 0
        actual = true[true['itemID']==itemID]['demandPrediction'].iloc[0]
        predicted = pred[pred['itemID']==itemID]['demandPrediction'].iloc[0]
        price = priceData[priceData['itemID']==itemID]['simulationPrice'].iloc[0]
        if actual >= predicted: 
            s += price * predicted
        else: 
            s += price * actual
            s -= 0.6 * price * (predicted - actual)
        if verbose > 0: 
            print('Score for item with ID {}:\t {:.2f} (max: {})'.format(itemID, s, price*actual))
        score_list.append(s)
    return score_list
```
